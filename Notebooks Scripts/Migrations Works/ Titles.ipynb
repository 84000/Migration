```python
# load works JSON file from a URL
import requests
import json

url = 'https://read.84000-translate.org/translations.json?api-version=0.3.0'
response = requests.get(url)
works = json.loads(response.text)
display(works)
```

```python
# Convert JSON to Pandas DataFrame
import pandas as pd

Works_table = pd.DataFrame(works['work'])
display(Works_table)
```

```python
# Table Definations
Works_table = pd.DataFrame(columns=['xmlid', 'title', 'url', 'type', 'author', 'translator', 'language', 'source', 'source_url', 'source
Title_table = pd.DataFrame(columns=['work_id', 'title', 'type'])
```

```python
# Sample Work loading Titles
import dotenv
import os
dotenv.load_dotenv()

header = {'Content-Type': 'application/json',
          'Authorization': dotenv.get_key(dotenv.find_dotenv(), 'SERVER_API_KEY')}

def load_titles(work_url):
    response = requests.get("https://read.84000-translate.org/"+work_url, headers=header)
    titles = json.loads(response.text)['title']
    return titles

display(load_titles(Works_table['url'][0]))
```

```python
current_work = Works_table.iloc[1]
curr_titles = load_titles(current_work['url'])

# add current work titles to Title_table
for title in curr_titles:
    Title_table = Title_table.append()
```
