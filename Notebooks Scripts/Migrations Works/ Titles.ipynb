```python
import os

# Table Definations
import pandas as pd

Works_table = pd.DataFrame(columns=['xmlId','url', 'workType', 'toh','worktype'])
```

```python
# load works JSON file from a URL
import requests
import json

url = 'https://read.84000-translate.org/translations.json?api-version=0.4.0'
response = requests.get(url)
works = json.loads(response.text)
display(works)
```

```python
# Convert JSON to Pandas DataFrame

for work in works['translation']:
    new_row = pd.DataFrame({'xmlId': [work['workId']],
                            'url': [work['url']],
                            'workType': [work['work-type']],
                            'toh': [work['toh']],
                            'worktype': [work['work-type']]})
    Works_table = pd.concat([Works_table, new_row], ignore_index=True)
display(Works_table)
```
