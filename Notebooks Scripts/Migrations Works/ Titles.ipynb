```python
import os

# Table Definations
import pandas as pd

Works_table = pd.DataFrame(columns=['xmlId','url', 'workType', 'toh','worktype'])
```

```python
# load works JSON file from a URL
import requests
import json

url = 'https://read.84000-translate.org/translations.json?api-version=0.4.0'
response = requests.get(url)
works = json.loads(response.text)
display(works)
```

```python
# Convert JSON to Pandas DataFrame

for work in works['translation']:
    new_row = pd.DataFrame({'xmlId': [work['workId']],
                            'url': [work['url']],
                            'workType': [work['work-type']],
                            'toh': [work['toh']],
                            'worktype': [work['work-type']]})
    Works_table = pd.concat([Works_table, new_row], ignore_index=True)
display(Works_table)
```

```python
# Sample Work loading Titles


def load_titles(Work_id):
    header = {'Content-Type': 'application/json',
          'Authorization': 'Basic YW1hbi1zaGFybWE6dm9IYlE3OGNZ'}
    response = requests.get(f"""https://read.84000-translate.org/translation/{Work_id}.json?api-version=0.4.0&annotate=false""", headers=header)
    titles = json.loads(response.text)
    return titles
```

```python
Workdata = load_titles(Works_table['url'][0])
```

```python
display(Workdata)
```

```python
# empty Titles_table
Titles_table = pd.DataFrame(columns=['work_id', 'label', 'titleType','content','language'])
```

```python
# Append Titles from work to Titles_table

for title in Workdata['title']:
    new_row = pd.DataFrame({'work_id': [Workdata['text-id']],
                            'label': '',
                            'titleType': '',
                            'content': [title['#text']],
                            'language': [title['xml:lang']]})
    Titles_table = pd.concat([Titles_table, new_row], ignore_index=True)

```

```python

display(Titles_table)
```

```python
# empty Passage_table
Passage_table = pd.DataFrame(columns=['work_id','content','segmentationType','xmlId','folio_id','folioPointer'])
```

```python

for passage in Workdata['translation']:
    new_row = pd.DataFrame({'work_id': [Workdata['text-id']],
                            'content': [passage['#text']],
                            'segmentationType': 'Passage',
                            'xmlId': passage['location-id'],
                            'folio_id': '',
                            'folioPointer': ''})
    Passage_table = pd.concat([Passage_table, new_row], ignore_index=True)
```

```python
display(Passage_table)
```
