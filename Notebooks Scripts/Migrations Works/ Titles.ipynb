```python
# load works JSON file from a URL
import requests
import json

url = 'https://read.84000-translate.org/translations.json?api-version=0.4.0'
response = requests.get(url)
works = json.loads(response.text)
display(works)
```

```python
Works_table = pd.DataFrame(columns=['xmlId','url', 'workType', 'toh','worktype'])

# Convert JSON to Pandas DataFrame
for work in works['work']:
    new_row = pd.DataFrame({'xmlId': [work['workId']],
                            'url': [work['url']],
                            'workType': [work['workType']],
                            'toh': [work['catalogueWorkIds']]})
    Works_table = pd.concat([Works_table, new_row], ignore_index=True)



#modify value of a workType
Works_table.loc[Works_table['workType'] == 'eft:translation', 'workType'] = 'Translation'
display(Works_table)
```

```python
from supabase import create_client,Client
import os
# get environment variables from .env file
import os

def load_env():
    with open('.env', 'r') as f:
        for line in f:
            if line.strip():
                key, value = line.strip().split('=', 1)
                os.environ[key] = value

load_env()


# get environment variables
url: str = os.getenv("SUPABASE_URL")
key: str = os.getenv("SUPABASE_KEY")
supabase = create_client(url, key)

```

```python
# Create Titles dict from work 
WorkId = 1
# display the title of the work
titles = Workdata['work'][0]['title']

titles_dicts = [dict({'work_id': WorkId,
                      'content': title['content'],
                      'titleType': title['titleType'],
                      'language': title['language'],
                      'titleMigrationId': title['titleMigrationId']}) for title in titles]
       
display(titles_dicts)
```

```python
# upsert into supabase titles table
data = []
try:
    data,count = supabase.table('titles').upsert(titles_dicts).execute()
    print("Added",data[1])
except Exception as e:
    print("Error Occured",e)
    # If error occurs delete all just inserted data
    try:
        if (data[1]):
            for inserts in data[1]:
                print(inserts)
                data_temp,count_temp = supabase.table('titles').delete().eq('id',inserts['id']).execute()
                print('deleted',data_temp[1])
    except IndexError:
        print("No Data to Delete")
```

```python
# Adding Passages to Passage Table with WorkID
passages = Workdata['translation']['passage']
for passage in passages:
    new = ({'work_id': WorkId,
            'content': passage['content'],
            'segmentationType': passage['segmentationType'],
            'xmlId': passage['xmlId']})
    # add this to try catch block
    try:
        data,count = supabase.table('passages').upsert([new], on_conflict='nothing').execute()
    except:
        print("Error Occured",data)
    print(data)
```
